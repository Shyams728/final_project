```python
# -*- coding: utf-8 -*-
"""Kannada MNIST Project - FIXED VERSION.ipynb

Fixed version with corrected data leakage issues and improved implementation.
"""

"""Kannda MNIST Classification project Description:

The project aims to perform multi-class classification on handwritten digits from the Kannada MNIST dataset using various classification algorithms such as PCA, including different numbers of PCA components such as 10, 20, 30, etc. Additionally, it will utilize classification algorithms such as Decision Tree Classifier, Random Forest Classifier, K-Nearest Neighbors Classifier, and Support Vector Machine Classifier. The goal is to analyze metrics like accuracy, precision, recall, and F1 score for each algorithm. Furthermore, the project will display the confusion matrix and the ROC curve.

# Kannda MNIST Classification Project

## Kannda MNIST Classification Exploration

#### Install required packages
"""

!pip install streamlit pyngrok mlflow scikit-learn numpy pandas matplotlib seaborn plotly joblib

"""### Understand the dataset

#### Load the Kannada MNIST dataset
"""

from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load training data
X_train = np.load("/content/drive/MyDrive/data/Kannada_MNIST_datataset_paper/Kannada_MNIST_npz/Kannada_MNIST/X_kannada_MNIST_train.npz")["arr_0"]
y_train = np.load("/content/drive/MyDrive/data/Kannada_MNIST_datataset_paper/Kannada_MNIST_npz/Kannada_MNIST/y_kannada_MNIST_train.npz")["arr_0"]

# Load testing data
X_test = np.load("/content/drive/MyDrive/data/Kannada_MNIST_datataset_paper/Kannada_MNIST_npz/Kannada_MNIST/X_kannada_MNIST_test.npz")["arr_0"]
y_test = np.load("/content/drive/MyDrive/data/Kannada_MNIST_datataset_paper/Kannada_MNIST_npz/Kannada_MNIST/y_kannada_MNIST_test.npz")["arr_0"]

"""#### Size of data & label"""

print(f"Training data shape: {X_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Testing data shape: {X_test.shape}")
print(f"Testing labels shape: {y_test.shape}")

"""#### Sample image view"""

import matplotlib.pyplot as plt

# Display sample images
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i in range(10):
    ax = axes[i//5, i%5]
    ax.imshow(X_train[i], cmap='gray')
    ax.set_title(f'Digit: {y_train[i]}')
    ax.axis('off')
plt.tight_layout()
plt.show()

# Check pixel intensity range
print(f"Min pixel intensity: {X_train.min()}")
print(f"Max pixel intensity: {X_train.max()}")

labels = np.unique(y_train)
print('Unique labels:', labels)

"""### Pre Process the dataset

#### Flatten images
"""

# Flatten the image data
X_train_reshaped = X_train.reshape(X_train.shape[0], -1)
X_test_reshaped = X_test.reshape(X_test.shape[0], -1)

print(f"Flattened training shape: {X_train_reshaped.shape}")
print(f"Flattened testing shape: {X_test_reshaped.shape}")

"""#### **CRITICAL FIX: Create proper validation split**"""

from sklearn.model_selection import train_test_split

# Create validation split from training data (80% train, 20% validation)
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train_reshaped, y_train, test_size=0.2, random_state=42, stratify=y_train
)

print(f"Training split: {X_train_split.shape}, {y_train_split.shape}")
print(f"Validation split: {X_val_split.shape}, {y_val_split.shape}")
print(f"Test set: {X_test_reshaped.shape}, {y_test.shape}")

"""#### Standardize the data"""

from sklearn.preprocessing import MinMaxScaler

# Initialize scaler
scaler = MinMaxScaler()

# Fit ONLY on training data
X_train_scaled = scaler.fit_transform(X_train_split)

# Transform validation and test data using the SAME scaler
X_val_scaled = scaler.transform(X_val_split)
X_test_scaled = scaler.transform(X_test_reshaped)

"""#### PCA

**CRITICAL FIX: Apply PCA correctly without data leakage**
"""

from sklearn.decomposition import PCA

# Apply PCA (fit on training, transform on all)
pca = PCA(n_components=60)
X_train_pca = pca.fit_transform(X_train_scaled)  # Fit on training
X_val_pca = pca.transform(X_val_scaled)          # Transform validation
X_test_pca = pca.transform(X_test_scaled)        # Transform test

# Calculate explained variance
evr = pca.explained_variance_ratio_
total_variance_explained = sum(evr) * 100
print(f"Total variance explained by 60 components: {total_variance_explained:.2f}%")

"""##### Visualization

###### Explained Variance Ratio
"""

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_), 'b-')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Explained Variance Ratio')
plt.grid(True)
plt.show()

# Find optimal number of components (95% variance)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Components needed for 95% variance: {n_components_95}")

"""###### 2D Visualization"""

import plotly.express as px

# Use a subset for faster visualization
sample_indices = np.random.choice(len(X_train_pca), 1000, replace=False)
fig = px.scatter(
    x=X_train_pca[sample_indices, 0],
    y=X_train_pca[sample_indices, 1],
    color=y_train_split[sample_indices].astype(str),
    title='PCA Visualization (First 2 Components)',
    labels={'x': 'First Principal Component', 'y': 'Second Principal Component'},
    color_discrete_sequence=px.colors.qualitative.Set1
)
fig.update_layout(width=700, height=500)
fig.show()

"""### Training the Model

#### Random Forest Model Training
"""

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Train model with validation
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_pca, y_train_split)

# Predict on validation set
y_val_pred = rf_model.predict(X_val_pca)
val_accuracy = accuracy_score(y_val_split, y_val_pred)
print(f"Validation Accuracy: {val_accuracy:.4f}")

# Predict on test set
y_test_pred = rf_model.predict(X_test_pca)
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f"Test Accuracy: {test_accuracy:.4f}")

"""#### Visualize predictions"""

# Visualize some predictions
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i in range(10):
    ax = axes[i//5, i%5]
    img_idx = i * 100  # Sample some test images
    ax.imshow(X_test[img_idx], cmap='gray')
    true_label = y_test[img_idx]
    pred_label = y_test_pred[img_idx]
    color = 'green' if true_label == pred_label else 'red'
    ax.set_title(f'True: {true_label}\nPred: {pred_label}', color=color)
    ax.axis('off')
plt.tight_layout()
plt.show()

"""#### Performance Metrics"""

from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Calculate metrics (using weighted average for multi-class)
precision = precision_score(y_test, y_test_pred, average='weighted')
recall = recall_score(y_test, y_test_pred, average='weighted')
f1 = f1_score(y_test, y_test_pred, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_test_pred))

"""##### Confusion Matrix"""

import seaborn as sns

conf_matrix = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=range(10), yticklabels=range(10))
plt.title('Confusion Matrix - Random Forest')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

"""##### ROC Curve"""

from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

# Get predicted probabilities
y_prob = rf_model.predict_proba(X_test_pca)

# Binarize the labels
y_test_bin = label_binarize(y_test, classes=range(10))

# Compute ROC curve and ROC area for each class
plt.figure(figsize=(10, 8))
for i in range(10):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'Digit {i} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Each Digit')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

"""#### Experiment with Different PCA Component Sizes"""

from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import time

pca_components = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 60]
results = []

for n_components in pca_components:
    # Apply PCA
    pca_temp = PCA(n_components=n_components)
    X_train_pca_temp = pca_temp.fit_transform(X_train_scaled)
    X_val_pca_temp = pca_temp.transform(X_val_scaled)

    # Train model
    start_time = time.time()
    model_temp = RandomForestClassifier(n_estimators=100, random_state=42)
    model_temp.fit(X_train_pca_temp, y_train_split)
    training_time = time.time() - start_time

    # Evaluate
    y_val_pred_temp = model_temp.predict(X_val_pca_temp)
    val_accuracy = accuracy_score(y_val_split, y_val_pred_temp)

    # Calculate variance explained
    total_var = sum(pca_temp.explained_variance_ratio_) * 100

    results.append({
        'n_components': n_components,
        'val_accuracy': val_accuracy,
        'training_time': training_time,
        'variance_explained': total_var
    })

    print(f"Components: {n_components:2d} | "
          f"Val Accuracy: {val_accuracy:.4f} | "
          f"Time: {training_time:.2f}s | "
          f"Variance: {total_var:.1f}%")

# Plot results
import pandas as pd
results_df = pd.DataFrame(results)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(results_df['n_components'], results_df['val_accuracy'], 'bo-')
ax1.set_xlabel('Number of PCA Components')
ax1.set_ylabel('Validation Accuracy')
ax1.set_title('Accuracy vs PCA Components')
ax1.grid(True)

ax2.plot(results_df['n_components'], results_df['variance_explained'], 'ro-')
ax2.set_xlabel('Number of PCA Components')
ax2.set_ylabel('Variance Explained (%)')
ax2.set_title('Variance Explained vs PCA Components')
ax2.grid(True)

plt.tight_layout()
plt.show()

"""### Build a Complete Pipeline"""

from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn import set_config

set_config(display='diagram')

# Create full pipeline
full_pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    ('pca', PCA()),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Define parameter grid
param_grid = {
    'pca__n_components': [10, 30, 50, 70],
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [None, 10, 20],
}

# Perform grid search
grid_search = GridSearchCV(
    full_pipeline,
    param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# Fit on training data
grid_search.fit(X_train_split, y_train_split)

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validation score: {grid_search.best_score_:.4f}")

# Evaluate on validation set
best_pipeline = grid_search.best_estimator_
y_val_pred_best = best_pipeline.predict(X_val_split)
val_accuracy_best = accuracy_score(y_val_split, y_val_pred_best)
print(f"Validation accuracy with best pipeline: {val_accuracy_best:.4f}")

# Visualize the pipeline
best_pipeline

"""### Multi-Model Comparison"""

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

# Define models to compare
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100),
    'Naive Bayes': GaussianNB(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Support Vector Machine': SVC(random_state=42, probability=True)
}

# Evaluate each model with cross-validation
results_comparison = []
for name, model in models.items():
    # Create pipeline for each model
    pipeline = Pipeline([
        ('scaler', MinMaxScaler()),
        ('pca', PCA(n_components=30)),  # Fixed PCA components for fair comparison
        ('classifier', model)
    ])

    # Use cross-validation
    cv_scores = cross_val_score(
        pipeline,
        X_train_split,
        y_train_split,
        cv=5,
        scoring='accuracy',
        n_jobs=-1
    )

    results_comparison.append({
        'Model': name,
        'Mean CV Accuracy': cv_scores.mean(),
        'Std CV Accuracy': cv_scores.std()
    })

# Display results
comparison_df = pd.DataFrame(results_comparison)
print("\nModel Comparison Results:")
print(comparison_df.sort_values('Mean CV Accuracy', ascending=False))

# Visualize comparison
plt.figure(figsize=(10, 6))
bars = plt.bar(comparison_df['Model'], comparison_df['Mean CV Accuracy'])
plt.errorbar(comparison_df['Model'], comparison_df['Mean CV Accuracy'],
             yerr=comparison_df['Std CV Accuracy'], fmt='none', color='black', capsize=5)
plt.ylabel('Cross-Validation Accuracy')
plt.title('Model Comparison')
plt.ylim(0.7, 1.0)
plt.xticks(rotation=45, ha='right')

# Add value labels on bars
for bar, value in zip(bars, comparison_df['Mean CV Accuracy']):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
             f'{value:.3f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

"""### Final Production Model"""

import joblib
import json
from datetime import datetime

# Train final model on full training data
final_pipeline = Pipeline([
    ('scaler', MinMaxScaler()),
    ('pca', PCA(n_components=60)),
    ('classifier', SVC(C=10, kernel='rbf', probability=True, random_state=42))
])

# Train on combined training + validation data
X_full_train = np.vstack([X_train_scaled, X_val_scaled])
y_full_train = np.concatenate([y_train_split, y_val_split])

final_pipeline.fit(X_full_train, y_full_train)

# Evaluate on test set
y_test_final_pred = final_pipeline.predict(X_test_scaled)
test_accuracy_final = accuracy_score(y_test, y_test_final_pred)

print("=" * 50)
print("FINAL MODEL EVALUATION")
print("=" * 50)
print(f"Test Accuracy: {test_accuracy_final:.4f}")
print(f"Test Precision: {precision_score(y_test, y_test_final_pred, average='weighted'):.4f}")
print(f"Test Recall: {recall_score(y_test, y_test_final_pred, average='weighted'):.4f}")
print(f"Test F1-Score: {f1_score(y_test, y_test_final_pred, average='weighted'):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_test_final_pred))

# Save the model
model_filename = 'kannada_mnist_final_model.pkl'
joblib.dump(final_pipeline, model_filename)

# Save metadata
metadata = {
    'model_type': 'SVC with PCA',
    'pca_components': 60,
    'classifier': 'SVC with C=10, kernel=rbf',
    'input_shape': (28, 28),
    'classes': list(range(10)),
    'test_accuracy': float(test_accuracy_final),
    'created_at': datetime.now().isoformat(),
    'dataset': 'Kannada MNIST'
}

with open('model_metadata.json', 'w') as f:
    json.dump(metadata, f, indent=2)

print(f"\nModel saved as: {model_filename}")
print("Metadata saved as: model_metadata.json")

"""### Streamlit Deployment App"""

%%writefile app.py
import streamlit as st
import numpy as np
import joblib
import json
from PIL import Image
import matplotlib.pyplot as plt

# Page configuration
st.set_page_config(
    page_title="Kannada MNIST Classifier",
    page_icon="🔢",
    layout="wide"
)

@st.cache_resource
def load_model():
    """Load the trained model and metadata."""
    try:
        model = joblib.load('kannada_mnist_final_model.pkl')
        with open('model_metadata.json', 'r') as f:
            metadata = json.load(f)
        return model, metadata
    except FileNotFoundError:
        st.error("Model files not found. Please train the model first.")
        return None, None
    except Exception as e:
        st.error(f"Error loading model: {str(e)}")
        return None, None

def preprocess_image(image):
    """Preprocess uploaded image for prediction."""
    try:
        # Convert to grayscale and resize
        img = image.convert('L').resize((28, 28))
        # Convert to numpy array and normalize
        img_array = np.array(img) / 255.0
        # Flatten
        img_flat = img_array.reshape(1, -1)
        return img_flat, img_array
    except Exception as e:
        st.error(f"Error processing image: {str(e)}")
        return None, None

def main():
    st.title("🔢 Kannada MNIST Digit Classifier")
    st.markdown("---")

    # Load model
    model, metadata = load_model()
    if model is None:
        return

    # Sidebar
    with st.sidebar:
        st.header("About")
        st.info(
            "This app classifies handwritten Kannada digits (0-9) using "
            "a Support Vector Machine model with PCA dimensionality reduction."
        )

        st.header("Model Information")
        if metadata:
            st.write(f"**Model Type:** {metadata['model_type']}")
            st.write(f"**PCA Components:** {metadata['pca_components']}")
            st.write(f"**Classifier:** {metadata['classifier']}")
            st.write(f"**Test Accuracy:** {metadata['test_accuracy']:.3f}")

        st.header("Instructions")
        st.write("1. Upload an image of a Kannada digit")
        st.write("2. The app will display the preprocessed image")
        st.write("3. View the prediction and confidence scores")

    # Main content
    col1, col2 = st.columns([1, 1])

    with col1:
        st.subheader("Upload Image")
        uploaded_file = st.file_uploader(
            "Choose an image file",
            type=["png", "jpg", "jpeg"],
            help="Upload an image of a handwritten Kannada digit (0-9)"
        )

        if uploaded_file is not None:
            # Load and display original image
            image = Image.open(uploaded_file)
            st.image(image, caption="Original Image", use_column_width=True)

    with col2:
        st.subheader("Prediction Results")

        if uploaded_file is not None:
            # Preprocess and predict
            with st.spinner('Processing image...'):
                img_flat, img_array = preprocess_image(image)

                if img_flat is not None:
                    # Make prediction
                    prediction = model.predict(img_flat)[0]
                    probabilities = model.predict_proba(img_flat)[0]

                    # Display preprocessed image
                    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

                    ax1.imshow(image, cmap='gray')
                    ax1.set_title("Original Image")
                    ax1.axis('off')

                    ax2.imshow(img_array, cmap='gray')
                    ax2.set_title("Preprocessed (28x28)")
                    ax2.axis('off')

                    st.pyplot(fig)

                    # Display prediction
                    st.success(f"**Predicted Digit: {prediction}**")

                    # Display confidence scores
                    st.subheader("Confidence Scores")

                    fig2, ax = plt.subplots(figsize=(10, 4))
                    bars = ax.bar(range(10), probabilities, color='skyblue')
                    bars[prediction].set_color('green')

                    ax.set_xlabel('Digit')
                    ax.set_ylabel('Probability')
                    ax.set_xticks(range(10))
                    ax.set_ylim(0, 1)
                    ax.grid(True, alpha=0.3)

                    # Add value labels
                    for i, (bar, prob) in enumerate(zip(bars, probabilities)):
                        height = bar.get_height()
                        ax.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                               f'{prob:.3f}', ha='center', va='bottom',
                               fontsize=9 if prob < 0.1 else 10)

                    st.pyplot(fig2)

                    # Show detailed probabilities
                    with st.expander("View Detailed Probabilities"):
                        prob_df = pd.DataFrame({
                            'Digit': range(10),
                            'Probability': probabilities
                        }).sort_values('Probability', ascending=False)
                        st.dataframe(prob_df.style.format({'Probability': '{:.4f}'}))
                else:
                    st.error("Failed to process the image.")
        else:
            st.info("Please upload an image to see predictions.")

            # Show sample digits
            st.subheader("Sample Kannada Digits")
            st.write("The model is trained to recognize these digits (0-9):")

            # Create a simple grid of digits 0-9
            cols = st.columns(5)
            for i in range(10):
                with cols[i % 5]:
                    st.markdown(f"**{i}**")
                    # You could add sample images here if available

if __name__ == '__main__':
    import pandas as pd  # Import here for the dataframe usage
    main()

```
